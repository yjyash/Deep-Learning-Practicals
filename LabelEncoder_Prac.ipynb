```python
df = pd.DataFrame(
    {'a' : ['Yes','no',12],
     'b' : ['Male','Female','Another']}
)
```

```python
df['values'] = 'Go' + df['b']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df['b'] = le.fit_transform(df['b'])
```

```python
df['values'] = le.fit_transform(df['values'])
```
