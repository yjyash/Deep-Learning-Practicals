```python
import keras
from keras import layers

input_img = keras.Input(shape=(28,28,1))

#Encoder
x = layers.Conv2D(16,3,activation="relu",strides=1,padding="same")(input_img)
x = layers.MaxPooling2D((2,2),padding="same")(x)
x = layers.Conv2D(8,3,activation="relu",strides=1,padding="same")(x)
x = layers.MaxPooling2D((2,2),padding="same")(x)
x = layers.Conv2D(8,3,activation="relu",strides=1,padding="same")(x)
x = layers.MaxPooling2D((2,2),padding="same")(x)
x = layers.Conv2D(8,3,activation="relu",strides=1,padding="same")(x)
encoder = layers.MaxPooling2D((2,2),padding="same")(x)


#Decoder
x = layers.Conv2D(8,3,activation="relu",strides=1,padding="same")(encoder)
x = layers.UpSampling2D((2,2))(x)
x = layers.Conv2D(8,3,activation="relu",strides=1,padding="same")(x)
x = layers.UpSampling2D((2,2))(x)
x = layers.Conv2D(8,3,activation="relu",strides=1,padding="same")(x)
x = layers.UpSampling2D((2,2))(x)
x = layers.Conv2D(16,3,activation="relu",strides=1,padding="same")(x)
x = layers.UpSampling2D((2,2))(x)
x = layers.Cropping2D(((2,2),(2,2)))(x)
decoder = layers.Conv2D(1,3,activation="sigmoid",padding="same")(x)
```

```python
autoencoder = keras.Model(input_img,decoder)
autoencoder.compile(optimizer="adam",loss="binary_crossentropy")
autoencoder.summary()
```

```python
from keras.datasets import mnist
import numpy as np
(x_train,_),(y_train,_)=mnist.load_data()
# print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
x_train = x_train.astype("float32")/255.
y_train = y_train.astype("float32")/255.
```

```python
x_train = np.reshape(x_train,(len(x_train),28,28,1))
y_train = np.reshape(y_train,(len(y_train),28,28,1))
```
