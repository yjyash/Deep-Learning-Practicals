```python
from keras.models import Model,Sequential
from keras.layers import Dense,Dropout,Conv2D,MaxPool2D
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen=ImageDataGenerator(rescale=1/255)
```

```python
train_generator = train_datagen.flow_from_directory(
        'C:\\Users\\yjyas\\CNN\\food20dataset\\train_set',
        target_size=(224, 224),
        batch_size=64)
```

```python
print(train_generator.class_indices)
```

```python
test_datagen=ImageDataGenerator(rescale=1/255)
```

```python
test_generator = test_datagen.flow_from_directory(
        'C:\\Users\\yjyas\\CNN\\food20dataset\\test_set',
        target_size=(224, 224),
        batch_size=64)
```

```python
print(test_generator.class_indices)
```

```python
test_generator.classes
```

```python
from keras.models import Model,Sequential
from keras.layers import Dense,Flatten,Conv2D,MaxPooling2D,Dropout
```

```python
model = Sequential()
model.add(Conv2D(filters = 64,kernel_size=(5,5),activation='relu',input_shape=(224,224,3), padding='Same'))
model.add(MaxPooling2D(pool_size=(3,3)))
model.add(Dropout(0.25))
model.add(Conv2D(filters = 128,kernel_size=(5,5),activation='relu',input_shape=(224,224,3),padding='Same'))
model.add(MaxPooling2D(pool_size=(3,3)))
model.add(Dropout(0.25))
model.add(Conv2D(filters = 128,kernel_size=(3,3),activation='relu',input_shape=(224,224,3),strides=1, padding='Same'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(256,activation='relu'))
# model.add(Dropout(0.2))
model.add(Dense(20,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(train_generator,epochs=50)
```

```python
pred1 = model.predict(test_generator)
```

```python
import numpy as np
for i in range(10):
    print(np.argmax(pred1[i]))
```
